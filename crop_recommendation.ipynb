```python
crop_encoding = {
    'Rice': 0, 'Maize': 1, 'Jute': 2, 'Cotton': 3, 'Coconut': 4,
    'Papaya': 5, 'Orange': 6, 'Apple': 7, 'Muskmelon': 8, 'Watermelon': 9,
    'Grapes': 10, 'Mango': 11, 'Banana': 12, 'Pomegranate': 13,
    'Lentil': 14, 'Blackgram': 15, 'MungBean': 16, 'MothBeans': 17,
    'PigeonPeas': 18, 'KidneyBeans': 19, 'ChickPea': 20, 'Coffee': 21
}

# Reverse mapping
reverse_crop_encoding = {v: k for k, v in crop_encoding.items()}
```

```python
def predict_crop(model, scaler):
    nitrogen = float(input("Enter Nitrogen level: "))
    phosphorus = float(input("Enter Phosphorus level: "))
    potassium = float(input("Enter Potassium level: "))
    temperature = float(input("Enter Temperature (¬∞C): "))
    humidity = float(input("Enter Humidity (%): "))
    ph = float(input("Enter pH Value: "))
    rainfall = float(input("Enter Rainfall (cm): "))

    # Create a DataFrame from input
    input_df = pd.DataFrame([[nitrogen, phosphorus, potassium, temperature, humidity, ph, rainfall]],
                            columns=['Nitrogen', 'Phosphorus', 'Potassium', 'Temperature', 'Humidity', 'pH_Value', 'Rainfall'])

    # Apply scaler
    input_scaled = scaler.transform(input_df)

    # Predict crop (returns encoded number)
    prediction_encoded = model.predict(input_scaled)[0]

    # Convert encoded prediction to crop name
    predicted_crop = reverse_crop_encoding[prediction_encoded]

    # Show result
    print(f"\nüåæ Recommended Crop: {predicted_crop}")
predict_crop(model,s)
```
